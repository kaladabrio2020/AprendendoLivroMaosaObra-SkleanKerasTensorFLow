```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(X,y,pred,xtest):
    plt.figure(figsize=(6,5))
    plt.scatter(X,y)
    plt.plot(xtest,pred,'*-',color='red')
    plt.grid(True)
    plt.ylabel('Valor Y(prever)')
    plt.xlabel('Caracteristica')
    plt.show()
```

### Regress√£o Polinomial

```python
X = 2*np.random.randn(100,1)
y = 0.5*X**2 + X+2 + np.random.randn(100,1)
```

```python
plt.scatter(X,y)
plt.savefig('polinomio.png')
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly  = PolynomialFeatures(degree=3,include_bias=False)
xpoly = poly.fit_transform(X)
print(xpoly[0],X[0],X[0]**2)
```

```python
from sklearn.linear_model import LinearRegression

linReg = LinearRegression()
linReg.fit(xpoly,y)

print(linReg.intercept_,linReg.coef_)
```

```python
def F(x,intercept,coef):
    fx = 0
    intercep = intercept[0]
    coef     = coef.tolist()[0]

    for i, coeficiente in enumerate(coef, start=1):
        fx += (coeficiente * x )** i
        
    return fx + intercep
```

```python
ye=F(X,linReg.intercept_,linReg.coef_)
```

```python
plt.scatter(X,y)
plt.plot(sorted(X),sorted(ye),color='red')
plt.show()
```
