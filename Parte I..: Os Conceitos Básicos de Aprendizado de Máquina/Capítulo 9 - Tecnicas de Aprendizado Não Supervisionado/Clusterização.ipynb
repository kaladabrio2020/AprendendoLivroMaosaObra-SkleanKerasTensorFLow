```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs
from sklearn.ensemble import *
from sklearn.model_selection import train_test_split
from sklearn.cluster  import KMeans
```

```python
X, y = make_blobs(n_samples=1000,n_features=2,centers=5,random_state=50,cluster_std=0.9)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
pred = kmeans.fit_predict(X)
```

```python
centerd = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=pred,cmap='viridis')
plt.scatter(centerd[:,0],centerd[:,1],c='red',label='Centroids')
plt.show()
```
